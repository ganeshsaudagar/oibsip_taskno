```python
#rename the columns name Region to State,Estimated Unemployment Rate (%) to Estimated Unemployment Rate,
# Estimated Labour Participation Rate (%) to Estimated Labour Participation Rate
df2.columns=['State', 'Date', 'Frequency', 'Estimated Unemployment Rate',
             'Estimated Employed', 'Estimated_Labour_Participation_Rate',
            'Area']
df2
```

```python
states=df2.groupby(["State",'Estimated Unemployment Rate']).size().reset_index()

states
```

Created new daatframe df3 which is showing state and their Region

```python
state_name=['Andhra Pradesh', 'Assam', 'Bihar', 'Chhattisgarh', 'Delhi', 'Goa',
       'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu & Kashmir',
       'Jharkhand', 'Karnataka', 'Kerala', 'Madhya Pradesh',
       'Maharashtra', 'Meghalaya', 'Odisha', 'Puducherry', 'Punjab',
       'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telangana', 'Tripura',
       'Uttar Pradesh', 'Uttarakhand', 'West Bengal', 'Chandigarh']
state_region=['Southern zone','North Estern Zone','North Estern Zone','Estern Zone','Central Zone','Nothern Zone','Southern Zone',"Western Zone",
             'Northern Zone','Northern Zone','Northern zone','Estern Zone','Southern Zone',"Southern Zone","Central Zone",
             'North Estern Zone',"Estern Zone","Southern Zone",'Northern Zone','Northern Zone',"Estern Zone",
             'Southern Zone','Southern Zone','North Estern Zone','Central Zone','Central Zone','Estern Zone','Central Zone']
df3=pd.DataFrame({"State":state_name,"Region":state_region})
df3
```

Merging df3 and df2 on State column so that to add Region column toprevious dataset

```python
df4=pd.merge(df3,df2,on='State',how='right')
df4
```

```python
df4['State'].unique()
```

```python
state_name=df4.State.value_counts().index
state_name
```

```python
state_values=df4.State.value_counts().values
state_values
```

# Pie chart showing top5 states in the Dataset

```python
plt.pie(state_values[:5],labels=state_name[:5],autopct="%1.2f%%")
```

```python
df4.columns
```

```python
data = df4.groupby(['State'])[['Estimated Unemployment Rate','Estimated Employed','Estimated_Labour_Participation_Rate']].mean()
data=pd.DataFrame(data).reset_index()
```

```python
data
```

```python
data_state=data["State"]
data_unemployment_rate=data['Estimated Unemployment Rate']
```

# BARPLOT SHOWING STATE WISE Estimated Unemploymnet Rate

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(14,6)
ax=sns.barplot(data=data,x=data_state[:],y=data_unemployment_rate[:])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.title("STATEWISE ESTIMATED UNEMPLOYED RATE")
plt.show()
```
